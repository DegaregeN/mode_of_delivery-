# Merging four different datasets

```python
def standardize_dataframe(df, source_name, all_columns):
    df = df.copy()   # ✅ FIX: avoid SettingWithCopyWarning
    
    df.columns = df.columns.str.strip()

    if 'column_mapping' in globals():
        df = df.rename(columns={
            k: v for k, v in column_mapping.items() if k in df.columns
        })

    for col in all_columns:
        if col not in df.columns:
            df[col] = np.nan

    df = df[all_columns]

    df["Institution"] = source_name  # now safe

    df = df.drop_duplicates()

    return df
```

```python
all_columns = list(
    set(data.columns)
    | set(data_G.columns)
    | set(data_Z.columns)
    | set(data_M.columns)
)
```

```python
data_Y = standardize_dataframe(data, 'Yekatit Hospital', all_columns)
data_G = standardize_dataframe(data_G, 'Ghandi Hospital', all_columns)
data_Z = standardize_dataframe(data_Z, 'Zewuditu Hospital', all_columns)
data_M = standardize_dataframe(data_M, 'Minilik Hospital', all_columns)
```

```python
df = pd.concat(
    [data_Y, data_G, data_M, data_Z],
    axis=0,
    ignore_index=True,
    verify_integrity=True  # Will raise error if duplicates still exist
)
```
